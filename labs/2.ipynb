## Краевая задача для обыкновенного дифференциального уравнения второго порядка

```python
import numpy as np
import matplotlib.pyplot as plt
from funcs import *

N = 100
x0, xN = 0, 1
h = (xN - x0) / N 
x = np.linspace(x0, xN, N + 1)

p = -1
q = -2
f = -3 * np.exp(-x)

r = -h / 2
a = (np.abs(r) + np.exp(-np.abs(r)) - r) / h**2
c = (np.abs(r) + np.exp(-np.abs(r)) + r) / h**2
b = -a - c + q

A = np.zeros((N + 1, N + 1))
F = np.zeros(N + 1)

for i in range(1, N):
    A[i, i - 1] = a
    A[i, i] = b
    A[i, i + 1] = c
    F[i] = f[i]

# Граничные условия
# u'(0) = 0 -> (u1 - u0) / h = 0
A[0, 0] = 1
A[0, 1] = -1
F[0] = 0

# u(1) + 2u'(1) = 0 -> uN + 2 (uN - uN-1) / h = 0
A[N, N] = 1 + 2 / h
A[N, N - 1] = -2 / h
F[N] = 0

U = np.linalg.solve(A, F)

U_exact = (x + 1) * np.exp(-x)

plt.figure(figsize=(10, 6))
plt.plot(x, U, "b-", label="Численное решение")
plt.plot(x, U_exact, "r-", label="Точное решение")
plt.xlabel("x")
plt.ylabel("u(x)")
plt.legend()
plt.title("Численное и точное решения")
plt.grid()
plt.show()
```
